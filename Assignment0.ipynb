6. Write a function definition named mode that takes in sequence of numbers and returns the most commonly occuring value

```python
def mode(seq):
  freq={} #empty dictionary {num:count}
  for num in seq:
    if num in freq:
      freq[num]+=1
    else:
        freq[num]=1
  #find max count in freq
  max_count=max(freq.values())
  print(max_count)
  # store nums of max count
  max_nums= []
  for num,count in freq.items():
   if count==max_count:
    max_nums.append(num)
  return max_nums

def main():
  n=int(input('Sequence length:'))
  seq=[]
  for i in range(1,n+1):
    num=int(input(f"Enter num {i}: "))
    seq.append(num)

  print("Highest freq num (Mode) is ", mode(seq))
main()
```

Exercise 1: Read Total profit of all months and show it using a line plot \\
Total profit data provided for each month. Generated line plot must include the following properties: â€“

X label name = Month Number \\
Y label name = Total profit \\




```python
data= pd.read_csv('/content/company_sales_data.csv')
data.head()
data['total_profit']
profit_data=data.sort_values(by='month_number', ascending=True)
profit_data['total_profit']
```

```python
plt.plot(profit_data['month_number'], profit_data['total_profit'], color='blue', marker='o')
plt.title('Total profit of all months')
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.show()
```
